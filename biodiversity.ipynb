# Introduction

This goal of this project is to analyze biodiversity data from the National Parks Service, particularly around various species observed in different national park locations.

This project will scope, analyze, prepare, plot data, and seek to explain the findings from the analysis.

Here are a few questions that this project has sought to answer:

- What is the distribution of conservation status for species?
- Are certain types of species more likely to be endangered?
- Are the differences between species and their conservation status significant?
- Which animal is most prevalent and what is their distribution amongst parks?

**Data sources:**

Both `Observations.csv` and `Species_info.csv` was provided by [Codecademy.com](https://www.codecademy.com).

Note: The data for this project is *inspired* by real data, but is mostly fictional.

```python
# Import Libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import CSV
obser = pd.read_csv("observations.csv")
spec = pd.read_csv("species_info.csv")

# print(obser.head())
print(len(obser.index))
# print(spec.head())
print(len(spec.index))
```

```python
# Drop duplicates rows
obser = obser.drop_duplicates()
spec = spec.drop_duplicates()

print(len(obser.index))
print(len(spec.index))

# Notice 15 rows in obser were removed due to being a duplicate
```

```python
# Which species were spotted the most at each park
ax = sns.catplot(data=obser, x="scientific_name", y="observations", hue="park_name", kind='bar', col='park_name')
ax.set_xticklabels(rotation=45, horizontalalignment='right')
plt.show()
plt.clf()
```

```python
# What is the distribution of conservation_status for animals?
sns.countplot(data= spec, x='category', hue ='conservation_status')
plt.show()
plt.clf()

# The graph breaks down the distibution of conservation_status of each Category of animal.
# The "Species of Concern" took the majority of the distribution from each category
# We can also see that Birds have the most cases of conservation_status
```

```python
# Are the differences between species and their conservation status significant?
sns.catplot(data = spec, x='conservation_status', hue='category', kind='count')
plt.show()
plt.clf()

# The Graph breaks down the conservation_status by each animal category.
# There is difference between species and their conservation status. In example Bird and Mammal has the most cases in the data provided. 
# 
```
